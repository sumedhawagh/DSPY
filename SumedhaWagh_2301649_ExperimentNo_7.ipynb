Q1. What is need for Hypothesis Testing?

Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. Since it is often difficult to collect data from the entire population, samples are used to test assumptions.

* To Make Decisions Using Sample Data:
It helps determine whether observed results are statistically significant or occurred by chance, allowing decisions about the population.

* To Test Assumptions or Claims:
Hypothesis testing involves two statements: the Null Hypothesis (H₀) and the Alternative Hypothesis (H₁). It helps decide whether to reject or accept the null hypothesis based on sample data.
Example: Testing whether the average age of students is 19 years.

* To Reduce Uncertainty:
It provides a scientific method to measure the reliability of results and minimizes errors in decision-making.

* To Compare Groups or Methods:
It is used to check whether there is a significant difference between two groups or techniques.
Example: Comparing teaching methods or medicines.

* To Support Research and Data Analysis:
It is widely used in statistics, data science, business analysis, and scientific research to validate results and draw meaningful conclusions.

Q2. With a neat diagram show important steps in hypothesis Testing

Steps in Hypothesis Testing:

1. Formulate Hypotheses
Define:
Null Hypothesis (H₀): No effect / no difference
Alternative Hypothesis (H₁): There is an effect / difference
2. Choose Significance Level (α)
Common values: 0.05 or 0.01
Represents the probability of making a Type I error
3. Select Appropriate Test
Based on data and problem:
z-test, t-test, chi-square test, ANOVA, etc.
4. Collect Sample Data & Compute Test Statistic
Gather sample data
Calculate test statistic (z, t, χ², etc.)
5. Determine Critical Region / p-value
Compare:
Test statistic with critical value, OR
p-value with α
6. Make Decision
If result falls in critical region or p-value < α → Reject H₀
Otherwise → Fail to reject H₀
7. Draw Conclusion
Interpret result in context of the problem
State whether there is sufficient evidence to support H₁

Q4. What is level of significance ? What role t plays in hypothesis Testing

**Definition**

The level of significance (α) is the probability of rejecting the null hypothesis (H₀) when it is actually true. In simple terms, it is the maximum risk of making a wrong decision (Type I error) that we are willing to accept.

**Common Values**

* α = 0.05 (5% risk) → most commonly used
* α = 0.01 (1% risk) → stricter testing

**Interpretation**

* α = 0.05 means: You accept a 5% chance of incorrectly rejecting a true H₀

**Role in Hypothesis Testing**

* Determines the critical region (rejection region)
* Used to compare with p-value
* If p-value < α → Reject H₀
* If p-value ≥ α → Fail to reject H₀

**Role of t in Hypothesis Testing**

t is the test statistic used in a t-test, which measures how far the sample mean is from the population mean in terms of standard error.

**Role of t**
1. Measures Difference
* Shows how much the sample result deviates from H₀
2. Decision Making Tool
* Compare calculated t with critical value from t-table
* OR compare p-value with α
3. Used When
* Sample size is small
* Population standard deviation is unknown
4. Helps Accept or Reject H₀
* Large t value → strong evidence against H₀
* Small t value → weak evidence against H₀

Q6. Check if you can accept Ho at alpha 95% ? Match your conclusions with python functions.

Step 1: Hypothesis Setup
H₀: μ = 19
H₁: μ ≠ 19 (two-tailed test)
α = 0.05 (since 95% confidence)

 Step 2: Decision Rule (Theory)

At α = 0.05 (two-tailed):

Reject H₀ if p-value < 0.05
Otherwise → Fail to reject H₀

Note: We don’t “accept” H₀, we say “fail to reject H₀”

Step 3: Example Data (Assume Sample)

Let’s assume you collected sample ages:

[18, 19, 20, 19, 18, 20, 19, 19, 18, 20]

Interpretation

Case 1: If output is like:

p-value = 0.32

 0.32 > 0.05

Fail to reject H₀

Conclusion: Average age can be considered 19

Case 2: If output is like:

p-value = 0.01

0.01 < 0.05
Reject H₀

Conclusion: Average age is NOT 19

At 95% confidence level (α = 0.05), we perform a two-tailed t-test. If the p-value is greater than 0.05, we fail to reject the null hypothesis, indicating that the average age can be considered 19. If the p-value is less than 0.05, we reject the null hypothesis. The same conclusion can be verified using Python’s scipy.stats.ttest_1samp() function.

```python
import numpy as np
from scipy import stats

# Sample data
data = [18, 19, 20, 19, 18, 20, 19, 19, 18, 20]

# Hypothesized mean
mu = 19

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(data, mu)

print("t-statistic:", t_stat)
print("p-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```
