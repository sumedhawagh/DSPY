**Q.2 In IRIS data set which non graphical EDA is performed ?**

Answer:Exploratory Data Analysis (EDA) is the process of examining a dataset to understand its structure, characteristics, and important patterns before applying any machine learning or data mining techniques. Non-graphical EDA refers to analyzing data using statistical summaries and data inspection methods without using charts or plots.

In the Iris dataset, the following non-graphical EDA techniques are performed:

1.Understanding the Structure of the Dataset (Shape of Data)
The first step in non-graphical EDA is to check the structure of the dataset. This includes identifying the number of rows (tuples) and columns (features).
The Iris dataset contains 150 records (tuples) and 5 attributes (features). Out of these, four attributes represent measurements of the flower and one attribute represents the species of the flower. This helps in understanding the size of the dataset and how the data is organized.

2.Identifying Feature Types (Data Types)
Another important step is identifying the type of each feature in the dataset. In the Iris dataset:

Sepal Length – Numerical

Sepal Width – Numerical

Petal Length – Numerical

Petal Width – Numerical

Species – Categorical (Nominal)

The first four attributes contain numerical values that represent measurements in centimeters, while the species attribute is categorical, indicating the type of Iris flower such as Setosa, Versicolor, and Virginica.

3.Checking for Missing Values
Non-graphical EDA also involves checking whether the dataset contains missing or null values. Missing values can affect analysis and model performance.
In the Iris dataset, no missing values are present, which means the dataset is complete and clean.

 4.Descriptive Statistical Summary
Descriptive statistics are used to summarize the main characteristics of numerical features. Common statistical measures include:

Mean – Average value of the feature

Median – Middle value when data is arranged in order

Minimum value – Smallest value in the dataset

Maximum value – Largest value in the dataset

Standard Deviation – Measures variation or spread of data

These statistics help understand the central tendency and variability of the numerical attributes such as sepal length, sepal width, petal length, and petal width.

5.Frequency Distribution of Categorical Attribute
For categorical attributes, non-graphical EDA includes calculating the frequency or count of each category.
In the Iris dataset, the species attribute has three classes:

Setosa – 50 records

Versicolor – 50 records

Virginica – 50 records

This shows that the dataset is balanced, meaning each class has the same number of observations.


6.Basic Data Inspection
Another step in non-graphical EDA is inspecting the first few rows of the dataset to understand how the data values are stored. This helps verify that the dataset has been loaded correctly and that the attributes contain appropriate values.

**Conclusion**:
Non-graphical EDA in the Iris dataset involves examining the dataset structure, identifying feature types, checking missing values, calculating descriptive statistics, and analyzing the frequency distribution of categorical variables. These techniques help in gaining a better understanding of the dataset before performing further data analysis or machine learning tasks.

**Q3.In Air Quality dataset which graph is used to perform multivariate analysis?**
Answer:In the Air Quality dataset, multivariate analysis is performed to study the relationship between multiple variables at the same time, such as temperature, humidity, CO levels, NOx levels, and other air pollutants.

The most commonly used graph for multivariate analysis in the Air Quality dataset is the Scatter Plot Matrix (Pair Plot).

A Scatter Plot Matrix displays scatter plots for every pair of variables in the dataset. Each plot shows the relationship between two variables, while the entire matrix helps analyze relationships among multiple variables simultaneously. This method helps in identifying correlation, patterns, and trends between different air quality parameters.

For example, in the Air Quality dataset, a scatter plot matrix can be used to observe relationships between variables like:

CO (Carbon Monoxide)

NOx (Nitrogen Oxides)

Temperature

Relative Humidity

Benzene concentration

By examining these scatter plots, we can understand how one pollutant or environmental factor affects another.

Thus, the Scatter Plot Matrix (Pair Plot) is commonly used for multivariate analysis in the Air Quality dataset because it allows simultaneous comparison of multiple variables and helps detect correlations and patterns in the data.
