**Q1. When do we go for Chi-square test?**

Answer: The Chi-square test is used when we want to analyze categorical data (data in the form of counts or frequencies). It helps in checking whether the observed data differs significantly from expected data.

We generally use the Chi-square test in the following situations:

When data is in frequency form (e.g., number of people, counts)

To test goodness of fit (whether observed data matches expected distribution)

To test independence between two categorical variables

To test homogeneity (whether different populations have the same distribution)

**Q2. Do we use Chi-square test for independence? Explain how.**

Answer:Yes, the Chi-square test is widely used to test independence between two categorical variables.

How it works:

```python
import numpy as np
from scipy.stats import chi2_contingency

# Contingency table
data = np.array([[10, 20, 30],
                 [6, 9, 17]])

chi2, p, dof, expected = chi2_contingency(data)

print("Chi-square value:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)

if p < 0.05:
    print("Reject H0: Variables are dependent")
else:
    print("Fail to reject H0: Variables are independent")
```

**Q4. Can we use Chi-square test to show existence of correlation between 2 features? How to interpret results?**

Answer: Yes, the Chi-square test can be used to check association (not exact correlation like Pearson) between two categorical variables.

Interpretation:

If p-value < 0.05:

There is a significant association between variables

This indicates presence of relationship (correlation in a broad sense)

If p-value ≥ 0.05:

No significant association

Variables are independent

Important Note:

Chi-square does not measure strength or direction of correlation

It only tells whether a relationship exists

Optional (Measure Strength):

You can use Cramér’s V:

```python
import numpy as np
from scipy.stats import chi2_contingency

data = np.array([[10, 20],
                 [20, 40]])

chi2, p, dof, expected = chi2_contingency(data)

n = data.sum()
cramers_v = np.sqrt(chi2 / (n * (min(data.shape)-1)))

print("Cramer's V:", cramers_v)
```
