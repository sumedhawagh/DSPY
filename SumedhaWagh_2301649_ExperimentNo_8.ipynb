Q1. What is 2 tailed and  tailed T Test ?

Two-Tailed and One-Tailed t-Test
1. One-Tailed t-Test

A one-tailed t-test is used when the research hypothesis predicts the direction of the effect (either greater than or less than a specific value).

It tests whether a parameter is either significantly higher OR significantly lower, but not both.
The entire level of significance (α) is placed in one tail of the distribution.



2. Two-Tailed t-Test

A two-tailed t-test is used when the research hypothesis does not specify direction, only that a difference exists.

It tests whether the parameter is significantly different (either higher or lower).
The level of significance (α) is split equally into both tails of the distribution.



Simple Interpretation

* Use one-tailed when you are confident about the direction of change.
* Use two-tailed when you only want to check if there is any difference at all.

Input and Output Characteristics of `scipy.stats.ttest_rel`

The function `scipy.stats.ttest_rel` is used to perform a paired (dependent) t-test, which compares the means of two related samples (for example, before–after measurements on the same subjects).

**Input Characteristics**

The function primarily takes two related datasets as input:

* `a` and `b` (arrays or lists):
These represent the two sets of observations being compared. Each value in a must correspond to a related value in b (same length and paired observations).
Example: marks of students before and after training.
* axis (optional):
Specifies the axis along which the test is performed when working with multi-dimensional data. By default, the test is applied along axis 0.
* nan_policy (optional):
Defines how missing values (NaN) are handled:
`propagate` → returns NaN if any NaN is present
`omit` → ignores NaN values
`raise` → throws an error
* alternative (optional):
Specifies the type of test:
`two-sided` → checks for any difference
`greater` → checks if first sample is greater
`less` → checks if first sample is smaller

Output Characteristics

The function returns a result object (usually called TtestResult) containing:

* `statistic` (t-value):
This is the calculated t-statistic, which measures the difference between paired samples relative to variability. A larger absolute value indicates a stronger difference.
* `pvalue`:

Q5. What is function in stats module for identifying power of a statistical test. Explain out put range of it and its significance.

Function for Identifying Power of a Statistical Test (in statsmodels)

In Python, the power of a statistical test is not directly provided in scipy.stats, but is computed using the statsmodels module. A commonly used function is:

* `statsmodels.stats.power.ttest_power()` (for one-sample or paired t-test)
* Similarly, `tt_ind_solve_power()` is used for independent samples.

Input Idea

To calculate power, the function typically requires:

* Effect size (magnitude of difference you expect)
* Sample size (n)
* Significance level (α)
* Type of test (one-tailed or two-tailed)

These inputs define how sensitive your test is.


Output Range

The output of the power function is a value between 0 and 1:

* Closer to 1 → High power (better ability to detect true effect)
* Closer to 0 → Low power (high chance of missing real effect)


Significance of Power

* Power represents the probability of correctly rejecting a false null hypothesis.

* It is equal to: Power=1−β
* A low power means your test might fail to detect real differences (high Type II error).
* A high power (≥ 0.8) is generally considered acceptable in practice.

Interpretation in Practice
* If power = 0.9 → 90% chance of detecting a true effect
* If power = 0.5 → Only 50% chance (unreliable test)

Conclusion

The power function in statsmodels helps you evaluate how effective your statistical test is before or after conducting it. It plays a critical role in experimental design, especially in deciding the required sample size and ensuring reliable conclusions.
