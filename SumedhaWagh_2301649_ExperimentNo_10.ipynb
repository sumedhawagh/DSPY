**Q1.Explain input and output characteristics for f_oneway function from stats library.**

Answer:1. Input and Output Characteristics of f_oneway Function (SciPy Stats Library)

The function f_oneway() is available in the scipy.stats module and is used to perform One-Way ANOVA (Analysis of Variance).

Syntax: from scipy.stats import f_oneway

f_statistic, p_value = f_oneway(sample1, sample2, sample3, ...)

**Input Characteristics:**

*Accepts two or more groups (samples) as input.

*Each input should be:

 *A list, array, or Pandas series
 *Numeric data (continuous values)

*Groups must be:

 *Independent

 *Normally distributed (approximately)

 *Have equal variances (homogeneity assumption)

 Example Input:

group1 = [10, 12, 14, 16]

group2 = [20, 22, 24, 26]

group3 = [30, 32, 34, 36]

**Output Characteristics:**

The function returns:

1.F-statistic

*Ratio of variance between groups to variance within groups

2.p-value

*Probability of observing the result under null hypothesis

Interpretation:

*If p-value < 0.05 → Reject Null Hypothesis (means differ)

*If p-value ≥ 0.05 → Fail to reject Null Hypothesis

```python
from scipy.stats import f_oneway

group1 = [10, 12, 14, 16]
group2 = [20, 22, 24, 26]
group3 = [30, 32, 34, 36]

f_stat, p_val = f_oneway(group1, group2, group3)

print("F-statistic:", f_stat)
print("P-value:", p_val)
```

**Q.2.Explain results of sm.stats.anova_lm fucntion.**

Answer:Explanation of Results of sm.stats.anova_lm() Function

The function anova_lm() is available in the statsmodels (sm) library and is used for performing ANOVA using linear models.

Syntax: import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('dependent_variable ~ independent_variable', data=df).fit()
anova_table = sm.stats.anova_lm(model)

Output Table Explanation:
| Column     | Meaning                    |
| ---------- | -------------------------- |
| **sum_sq** | Sum of Squares (variation) |
| **df**     | Degrees of Freedom         |
| **F**      | F-statistic value          |
| **PR(>F)** | p-value                    |


```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Create dataset
data = {
    'score': [85, 90, 88, 75, 78, 74, 92, 95, 94],
    'method': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C']
}

df = pd.DataFrame(data)

# Fit model
model = ols('score ~ method', data=df).fit()

# Perform ANOVA
anova_table = sm.stats.anova_lm(model)

print(anova_table)
```

Interpretation:

*High F value → Significant difference between groups

*Low p-value (<0.05) → Reject null hypothesis

*Helps determine if group means are significantly different

**Q4.When to use one way ANOVA? Explain with real world case.**

Answer:**When to Use One-Way ANOVA**

One-Way ANOVA is used when:

1.There is one independent variable (factor)

2.There are two or more groups

3.One dependent variable (numeric)

**Conditions to Use:**

1.Independent groups

2.Normally distributed data

3.Equal variances

**Real-World Case:**

**Problem:**

A college wants to check whether **different teaching methods** affect student performance.


1.Method A → Traditional

2.Method B → Online

3.Method C → Hybrid

Hypothesis:

H₀ (Null):All means are equal

H₁ (Alternative): At least one mean differs

```python
from scipy.stats import f_oneway

method_A = [85, 88, 90]
method_B = [75, 78, 74]
method_C = [92, 95, 94]

f_stat, p_val = f_oneway(method_A, method_B, method_C)

print("F-statistic:", f_stat)
print("P-value:", p_val)
```

**Conclusion:**

If p < 0.05 → Teaching method impacts performance

**Q5. When to use 2 way ANOVA ? Explain with real world case.?**

Answer:When to Use Two-Way ANOVA
Definition:

Two-Way ANOVA is used when:

1.There are two independent variables (factors)

2.One dependent variable

Example Factors:

Teaching Method (A, B, C)

Gender (Male, Female)

**Real-World Case:**

A company wants to analyze:

Does **advertisement type** and **region** affect sales?

Factors:

Advertisement: TV, Social Media

Region: Urban, Rural

**Two-Way ANOVA Model:**

model = ols('sales ~ C(ad_type) + C(region) + C(ad_type):C(region)', data=df).fit()
anova_table = sm.stats.anova_lm(model)

**Interpretation:**

*Main effects:

  Advertisement type effect

   Region effect

*Interaction effect:

Combined impact of both variables


```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

data = {
    'sales': [100, 120, 130, 150, 170, 180],
    'ad_type': ['TV', 'TV', 'Social', 'Social', 'TV', 'Social'],
    'region': ['Urban', 'Rural', 'Urban', 'Rural', 'Urban', 'Rural']
}

df = pd.DataFrame(data)

model = ols('sales ~ C(ad_type) + C(region) + C(ad_type):C(region)', data=df).fit()

anova_table = sm.stats.anova_lm(model)

print(anova_table)
```

**Conclusion:**

Helps understand **individual and combined effects**

Useful in complex real-world decision-making.
